```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf', 'svg')


sns.set_style('whitegrid')
sns.set_context('notebook')
sns.set_palette('Set1', 2)

def jukes_cantor(dist):
    return -(3/4) * log(1 - (4/3)*dist)

diffs = np.linspace(0, 0.7499, 750)
lens = [jukes_cantor(p) for p in diffs]

d = np.linspace(0, 1, 10)
```

```python
plt.plot(d, d, '--')
plt.plot(lens, diffs)
plt.xlim(0,1)
plt.ylim(0,1)
plt.ylabel('Obervable substutitions per site')
plt.xlabel('Actual substutitions per site')
plt.savefig('jukes-cantor.pdf')
```
